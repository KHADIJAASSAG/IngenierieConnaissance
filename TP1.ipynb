```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Univers de discours pour la température et la puissance de chauffage
x_temp = np.arange(0, 51, 1)  # Température de 0 à 50°C
x_chauffage = np.arange(0, 101, 1)  # Puissance de chauffage de 0 à 100%

# Définir les fonctions d'appartenance pour la température
froid = fuzz.trapmf(x_temp, [0, 0, 10, 20])
moyen = fuzz.trapmf(x_temp, [15, 20, 30, 35])
chaud = fuzz.trapmf(x_temp, [30, 40, 50, 50])

# Définir les fonctions d'appartenance pour le chauffage
faible = fuzz.trapmf(x_chauffage, [0, 0, 30, 50])
moyenne = fuzz.trapmf(x_chauffage, [30, 50, 70, 80])
forte = fuzz.trapmf(x_chauffage, [70, 90, 100, 100])

# Définir la température d'entrée
temp_input = 25  # Exemple de température d'entrée

# Calculer le degré d'appartenance pour l'entrée
temp_froid = fuzz.interp_membership(x_temp, froid, temp_input)
temp_moyen = fuzz.interp_membership(x_temp, moyen, temp_input)
temp_chaud = fuzz.interp_membership(x_temp, chaud, temp_input)

# Appliquer les règles floues
# Règles:
# 1. Si la température est Froid, alors le chauffage est Forte
# 2. Si la température est Moyen, alors le chauffage est Moyenne
# 3. Si la température est Chaud, alors le chauffage est Faible

# Déterminer les valeurs de chauffage à partir des règles
chauffage_forte = temp_froid * forte
chauffage_moyenne = temp_moyen * moyenne
chauffage_faible = temp_chaud * faible

# Combiner les sorties
aggregated = np.fmax(chauffage_forte, np.fmax(chauffage_moyenne, chauffage_faible))

# Calculer la sortie finale en utilisant le défuzzificateur
result = fuzz.defuzz(x_chauffage, aggregated, 'centroid')
print(f"Puissance du chauffage : {result:.2f}%")
```

- Explication des règles des règles
  - Règle 1 : Si la température est Froid, alors la puissance du chauffage est Forte.
  - Règle 2 : Si la température est Moyenne, alors la puissance du chauffage est Moyenne.
  - Règle 3 : Si la température est Chaud, alors la puissance du chauffage est Faible.
- Application des Règles :

  - Chaque règle est appliquée en multipliant le degré d'appartenance à la température par la fonction d'appartenance correspondante à la puissance de chauffage.
  - Les résultats sont ensuite combinés pour obtenir une sortie agrégée.
- Défuzzification :

  - La sortie finale est calculée en utilisant la méthode de défuzzification du centroid, qui donne un seul chiffre représentant la puissance de chauffage à appliquer.

```python
# Visualisation des fonctions d'appartenance et des résultats
plt.figure(figsize=(12, 8))

# Visualiser les fonctions d'appartenance de la température
plt.subplot(2, 1, 1)
plt.plot(x_temp, froid, 'b', label='Froid', linewidth=1.5)
plt.plot(x_temp, moyen, 'g', label='Moyen', linewidth=1.5)
plt.plot(x_temp, chaud, 'r', label='Chaud', linewidth=1.5)
plt.title('Fonctions d\'appartenance de la Température')
plt.xlabel('Température (°C)')
plt.ylabel('Degré d\'appartenance')
plt.legend()

# Visualiser les fonctions d'appartenance de chauffage
plt.subplot(2, 1, 2)
plt.plot(x_chauffage, faible, 'b', label='Faible', linewidth=1.5)
plt.plot(x_chauffage, moyenne, 'g', label='Moyenne', linewidth=1.5)
plt.plot(x_chauffage, forte, 'r', label='Forte', linewidth=1.5)
plt.fill_between(x_chauffage, aggregated, color='yellow', alpha=0.5, label='Sortie agrégée')
plt.axvline(result, color='k', linestyle='--', label='Sortie défuzzifiée')
plt.title('Fonctions d\'appartenance de la Puissance de Chauffage')
plt.xlabel('Puissance de chauffage (%)')
plt.ylabel('Degré d\'appartenance')
plt.legend()

plt.tight_layout()
plt.show()
```

#### 1- Créer les variables d’entrée et de sortie
- Variables d'entrée :

  - Distance : Distance à l'obstacle
  - Visibilité : Niveau de visibilité 
- Variable de sortie :

  - Vitesse : Vitesse recommandée du véhicule (exprimée en km/h)

```python
x_distance = np.arange(0, 101, 1)  # Distance de 0 à 100 m
x_visibilite = np.arange(0, 101, 1)  # Visibilité de 0 à 100 m
x_vitesse = np.arange(0, 121, 1)  # Vitesse de 0 à 120 km/h
```

#### 2- Les fonctions d'appartenance : 

```python
# Fonctions d'appartenance pour la distance
distance_proche = fuzz.trapmf(x_distance, [0, 0, 20, 30])
distance_moyenne = fuzz.trapmf(x_distance, [25, 35, 50, 60])
distance_loin = fuzz.trapmf(x_distance, [55, 70, 100, 100])

# Fonctions d'appartenance pour la visibilité
visibilite_bonne = fuzz.trapmf(x_visibilite, [60, 80, 100, 100])
visibilite_moyenne = fuzz.trapmf(x_visibilite, [30, 40, 60, 70])
visibilite_faible = fuzz.trapmf(x_visibilite, [0, 0, 10, 20])

# Fonctions d'appartenance pour la vitesse
vitesse_basse = fuzz.trapmf(x_vitesse, [0, 0, 30, 50])
vitesse_moyenne = fuzz.trapmf(x_vitesse, [40, 60, 80, 90])
vitesse_haute = fuzz.trapmf(x_vitesse, [80, 100, 120, 120])
```

#### 3- Les règles floues pour contrôler la vitesse en fonction des deux entrées. <br>
- Si la distance est proche ET la visibilité est faible, ALORS la vitesse est basse.
- Si la distance est proche ET la visibilité est moyenne, ALORS la vitesse est moyenne.
- Si la distance est proche ET la visibilité est bonne, ALORS la vitesse est moyenne.
- Si la distance est moyenne ET la visibilité est faible, ALORS la vitesse est basse.
- Si la distance est moyenne ET la visibilité est moyenne, ALORS la vitesse est moyenne.
- Si la distance est moyenne ET la visibilité est bonne, ALORS la vitesse est élevée.
- Si la distance est loin ET la visibilité est faible, ALORS la vitesse est basse.
- Si la distance est loin ET la visibilité est moyenne, ALORS la vitesse est moyenne.
- Si la distance est loin ET la visibilité est bonne, ALORS la vitesse est élevée.
- Si la visibilité est très faible, ALORS la vitesse doit être basse, peu importe la distance.

```python
# Règles floues
rule1 = fuzz.relation_product(distance_proche, vitesse_basse)
rule2 = fuzz.relation_product(distance_proche, vitesse_moyenne)
rule3 = fuzz.relation_product(distance_moyenne, vitesse_moyenne)
rule4 = fuzz.relation_product(distance_moyenne, vitesse_haute)
rule5 = fuzz.relation_product(distance_loin, vitesse_haute)
rule6 = fuzz.relation_product(visibilite_faible, vitesse_basse)
```

#### 4- Simulation des résultats 

```python
# Entrées
distance_input = 30  # Distance à l'obstacle
visibilite_input = 50  # Visibilité

# Calculer le degré d'appartenance pour les entrées
dist_proche = fuzz.interp_membership(x_distance, distance_proche, distance_input)
dist_moyenne = fuzz.interp_membership(x_distance, distance_moyenne, distance_input)
dist_loin = fuzz.interp_membership(x_distance, distance_loin, distance_input)

vis_faible = fuzz.interp_membership(x_visibilite, visibilite_faible, visibilite_input)
vis_moyenne = fuzz.interp_membership(x_visibilite, visibilite_moyenne, visibilite_input)
vis_bonne = fuzz.interp_membership(x_visibilite, visibilite_bonne, visibilite_input)

# Appliquer les règles floues
vitesse_basse_activation = np.fmax(dist_proche * vis_faible, vis_faible * vitesse_basse)
vitesse_moyenne_activation = np.fmax(dist_moyenne * vis_moyenne, dist_proche * vitesse_moyenne)
vitesse_haute_activation = np.fmax(dist_loin * vis_bonne, dist_moyenne * vitesse_haute)

# Combiner les sorties
aggregated = np.fmax(vitesse_basse_activation, np.fmax(vitesse_moyenne_activation, vitesse_haute_activation))

# Calculer la sortie finale en utilisant le défuzzificateur
vitesse_result = fuzz.defuzz(x_vitesse, aggregated, 'centroid')
print(f"Vitesse recommandée : {vitesse_result:.2f} km/h")
```
